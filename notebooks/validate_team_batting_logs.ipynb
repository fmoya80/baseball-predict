```python
import pandas as pd
```

```python
path = r"C:\Users\felip\OneDrive\Belgeler\Pipe\baseball-predict\data\interim\team_batting_logs_2026-03-21_to_2026-04-12.csv"
df = pd.read_csv(path)
df["game_date"] = pd.to_datetime(df["game_date"])
df.head()
```

```python
team_df = (
    df[df["team_name"] == "New York Yankees"]
    .sort_values(["game_date", "gamePk"])
    .reset_index(drop=True)
)

team_df[[
    "game_date",
    "gamePk",
    "team_name",
    "opponent_team_name",
    "hits",
    "at_bats",
    "walks",
    "hit_by_pitch",
    "sac_flies",
    "total_bases",
    "avg_game_last_5_avg",
    "obp_game_last_5_avg",
    "slg_game_last_5_avg",
    "ops_game_last_5_avg",
    "iso_game_last_5_avg",
]].tail(10)
```

```python
target_date = pd.Timestamp("2026-04-10")

target_row = team_df[team_df["game_date"] == target_date].copy()
target_row[[
    "game_date",
    "team_name",
    "avg_game_last_5_avg",
    "obp_game_last_5_avg",
    "slg_game_last_5_avg",
    "ops_game_last_5_avg",
    "iso_game_last_5_avg",
]]
```

```python
previous_games = (
    team_df[team_df["game_date"] < target_date]
    .sort_values(["game_date", "gamePk"])
    .tail(5)
    .copy()
)

previous_games[[
    "game_date",
    "gamePk",
    "opponent_team_name",
    "hits",
    "at_bats",
    "walks",
    "hit_by_pitch",
    "sac_flies",
    "total_bases",
    "avg_game",
    "obp_game",
    "slg_game",
    "ops_game",
]]
```

```python
hits_sum = previous_games["hits"].sum()
at_bats_sum = previous_games["at_bats"].sum()
walks_sum = previous_games["walks"].sum()
hbp_sum = previous_games["hit_by_pitch"].sum()
sf_sum = previous_games["sac_flies"].sum()
tb_sum = previous_games["total_bases"].sum()

avg_manual = hits_sum / at_bats_sum if at_bats_sum != 0 else None
obp_manual = (hits_sum + walks_sum + hbp_sum) / (at_bats_sum + walks_sum + hbp_sum + sf_sum) if (at_bats_sum + walks_sum + hbp_sum + sf_sum) != 0 else None
slg_manual = tb_sum / at_bats_sum if at_bats_sum != 0 else None
ops_manual = (obp_manual + slg_manual) if (obp_manual is not None and slg_manual is not None) else None
iso_manual = (slg_manual - avg_manual) if (slg_manual is not None and avg_manual is not None) else None

manual_check = pd.DataFrame([{
    "hits_sum": hits_sum,
    "at_bats_sum": at_bats_sum,
    "walks_sum": walks_sum,
    "hit_by_pitch_sum": hbp_sum,
    "sac_flies_sum": sf_sum,
    "total_bases_sum": tb_sum,
    "avg_manual": round(avg_manual, 3) if avg_manual is not None else None,
    "obp_manual": round(obp_manual, 3) if obp_manual is not None else None,
    "slg_manual": round(slg_manual, 3) if slg_manual is not None else None,
    "ops_manual": round(ops_manual, 3) if ops_manual is not None else None,
    "iso_manual": round(iso_manual, 3) if iso_manual is not None else None,
}])

manual_check
```

```python
comparison = pd.DataFrame([{
    "avg_dataset": target_row["avg_game_last_5_avg"].iloc[0],
    "obp_dataset": target_row["obp_game_last_5_avg"].iloc[0],
    "slg_dataset": target_row["slg_game_last_5_avg"].iloc[0],
    "ops_dataset": target_row["ops_game_last_5_avg"].iloc[0],
    "iso_dataset": target_row["iso_game_last_5_avg"].iloc[0],
}])

comparison
```
